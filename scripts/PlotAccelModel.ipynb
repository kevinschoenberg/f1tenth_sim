```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python

mu = np.linspace(0.1,1.,10)
print(mu)
x = np.linspace(0.00000001,7.51,100)
f = 0.3 * x * np.exp(1-(1-mu[8])*x)
a_optimal = np.array([1.11, 1.23, 1.34, 1.67, 2.0, 2.5, 3.33, 5.0, 7.51, 7.51])

sns.set_palette(sns.color_palette('coolwarm', 10))

for n in range(len(mu)):
  f = x
  plt.plot(x,f, label='mu = ' + str(np.round(n * 0.1 + 0.1, 1)))
  plt.legend()
  plt.axis('equal')
  plt.scatter(7.51, 7.51)
plt.xlabel('Desired acceleration m/s^2')
plt.ylabel('Actual acceleration m/s^2')
plt.title("Original acceleration model")
plt.savefig("original.pdf")
plt.show()

for n in range(len(mu)):
  f = x * mu[n]
  plt.plot(x,f, label='mu = ' + str(np.round(n * 0.1 + 0.1, 1)))
  plt.legend()
  plt.axis('equal')
  plt.scatter(7.51, 7.51 * mu[n])
plt.xlabel('Desired acceleration m/s^2')
plt.ylabel('Actual acceleration m/s^2')
plt.title("Acceleration model using coefficient of friciton")
plt.savefig("mu.pdf")
plt.show()

for n in range(len(mu)):
  f = 0.3 * x * np.exp(1-(1-mu[n])*x)
  plt.plot(x,f, label='mu = ' + str(np.round(n * 0.1 + 0.1, 1)))
  plt.legend()
  plt.axis('equal')
  plt.scatter(a_optimal[n], 0.3 * a_optimal[n] * np.exp(1-(1-mu[n])*a_optimal[n]))
plt.xlabel('Desired acceleration m/s^2')
plt.ylabel('Actual acceleration m/s^2')
plt.title("Acceleration model using acceleration coefficient of friciton")
plt.savefig("mu_a.pdf")
plt.show()


models = np.array([0.2, 0.4, 0.5, 0.6, 0.7, 0.8, 0.9])
a_models = np.array([1.26, 1.67, 2.0, 2.5, 3.33, 5.0, 7.51])

sns.set_palette(sns.color_palette('coolwarm', 7))

for n in range(len(models)):
  f = 0.3 * x * np.exp(1-(1-models[n])*x)
  plt.plot(x,f, label='mu = ' + str(models[n]))
  plt.legend()
  plt.axis('equal')
  plt.scatter(a_models[n], 0.3 * a_models[n] * np.exp(1-(1-models[n])*a_models[n]))
plt.xlabel('Desired acceleration m/s^2')
plt.ylabel('Actual acceleration m/s^2')
plt.title("Acceleration model using acceleration coefficient of friciton")
plt.savefig("models.pdf")
plt.show()



#for n in range(len(mu)):
  #f = 0.3 * x * np.exp(1-(1-mu[n])*x)
  #g = 1-f/(x)
  #plt.plot(g,f,label='loss_ratio ' + str(n))
  #plt.legend()
#plt.show() 
```

```python
v = np.linspace(1, 7, 7)
vm = np.linspace(0,7,1000)
print(v)

for n in range(len(v)):
  f = 2 * (v[n] - vm) * ((2.5)/7)
  for i in range(len(f)):
    if(f[i] > 2.5):
      f[i] = 2.5
  plt.plot(vm,f, label='desired vel = ' + str(n * 1 + 1))
  plt.legend()
  plt.axis('equal')
  plt.xlim(0,7)
  plt.ylim(0,5.2)
  plt.xlabel('Measured velocity m/s')
  plt.ylabel('Desired acceleration (max 2.5) m/s^2')
plt.title("Desired acceleration for different desired velocities")
plt.savefig("desired_acc_drop.pdf")
plt.show()

```
